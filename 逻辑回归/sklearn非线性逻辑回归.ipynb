```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import make_gaussian_quantiles #用于自己生成数据
from sklearn.preprocessing import PolynomialFeatures
```

```python
#生成2维正态分布，生成的数据按分位数分为两累，300个样本，2个特征样本
#可以生成两类或多类样本
x_data,y_data = make_gaussian_quantiles(n_samples=500,n_features=2,n_classes=2)
plt.scatter(x_data[:,0],x_data[:,1],c= y_data)
plt.show()
```

```python
logistic = linear_model.LogisticRegression()
logistic.fit(x_data,y_data)
```

```python
#获取数据值所在的范围
x_min,x_max = x_data[:,0].min()-1,x_data[:,0].max()+1
y_min,y_max = x_data[:,1].min()-1,x_data[:,1].max()+1

#生成网格矩阵
xx,yy = np.meshgrid(np.arange(x_min,x_max,0.02),
                   np.arange(y_min,y_max,0.02))
z = logistic.predict(np.c_[xx.ravel(),yy.ravel()])
#ravel和flatten 类似 多维转一维 flatten不改变原数据 ravel改变
# np.r_按row来组合array， 
# np.c_按colunm来组合array
# >>> a = np.array([1,2,3])
# >>> b = np.array([5,2,5])
# >>> np.r_[a,b]
# array([1, 2, 3, 5, 2, 5])
# >>> np.c_[a,b]
# array([[1, 5],
#        [2, 2],
#        [3, 5]])
# >>> np.c_[a,[0,0,0],b]
# array([[1, 0, 5],
#        [2, 0, 2],
#        [3, 0, 5]])
z = z.reshape(xx.shape)

#等高线
cs = plt.contourf(xx,yy,z)
plt.scatter(x_data[:,0],x_data[:,1],c= y_data)
plt.show()
print('score',logistic.score(x_data,y_data))
```
