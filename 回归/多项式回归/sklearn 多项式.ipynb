```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures #多项式
from sklearn.linear_model import LinearRegression
```

```python
data = np.genfromtxt("job.csv",delimiter = ",")
x_data = data[1:,1]
y_data = data[1:,2]
print(x_data)
plt.scatter(x_data,y_data)
plt.show()
```

```python
x_data = x_data[:,np.newaxis]
y_data = y_data[:,np.newaxis]
model = LinearRegression()
model.fit(x_data,y_data)
```

```python
plt.plot(x_data,y_data,'b.')
plt.plot(x_data,model.predict(x_data),'r')
plt.show()
```

```python
#定义多项式回归 ，degree 可调节多项式特征
poly_reg = PolynomialFeatures(degree = 7)
#特征处理
x_poly = poly_reg.fit_transform(x_data)
#定义回归模型
lin_reg = LinearRegression()
lin_reg.fit(x_poly,y_data)
```

```python
plt.plot(x_data,y_data,'b.')
x_test = np.linspace(1,10,100)[:,np.newaxis]#1~10 生成100个数据
plt.plot(x_test,lin_reg.predict(poly_reg.fit_transform(x_test)),'r')
plt.title("predict line")
plt.xlabel('job')
plt.ylabel('salary')
plt.show()
```
